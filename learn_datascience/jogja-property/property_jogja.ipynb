```python
import pandas as pd

# Load the data
df = pd.read_csv('data_cleaned.csv')

# Display the first few rows
df.head()
```

```python
# List of kabupaten
kabupaten_list = ["sleman", "bantul", "yogyakarta", "kulon progo", "gunung kidul"]

# Modified function to split location
def split_location_corrected_v4(location):
    # Split the location string using ','
    parts = [part.strip() for part in location.split(',')]
    
    # Initialize default values
    street, kapanewon, kabupaten, province = "N/A", "N/A", "N/A", "diy"
    
    # Assign values based on the kabupaten_list and given rules
    for part in reversed(parts):
        if any(sub_str in part for sub_str in ["jl.", "jalan", "jln"]):
            street = part
        elif part in kabupaten_list:
            kabupaten = part
        elif "di yogyakarta" not in part and (not kapanewon or kapanewon == "N/A"):
            kapanewon = part

    return street, kapanewon, kabupaten, province

# Apply the corrected function to the 'Location' column and create new columns
df['Street'], df['Kapanewon'], df['Kabupaten'], df['Province'] = zip(*df['Location'].apply(split_location_corrected_v4))
df.to_csv('processed_data.csv', index=False)

# Display the updated DataFrame
df.head(10)
```

```python
# Load the processed data
df_processed = pd.read_csv('processed_data.csv')

# Function to assign tags based on the title
def assign_tags(title):
    if "rumah" in title:
        return "house"
    elif "tanah" in title:
        return "land"
    elif "kost" in title:
        return "rent room"
    else:
        return "N/A"  # Default tag if none of the keywords are found

# Apply the function to the 'Title' column to create a new 'Tags' column
df_processed['Tags'] = df_processed['Title'].apply(assign_tags)

# Display the updated DataFrame with tags
df_processed.head(10)
```

```python
# Count the number of each tag in the 'Tags' column
tags_count = df_processed['Tags'].value_counts()

tags_count
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="geoapiExercises")

def geocode_location(location_name):
    location = geolocator.geocode(location_name + ", Yogyakarta, Indonesia")
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Get unique Kapanewon names
unique_kapanewon = df['Kapanewon'].unique()

# Geocode each unique Kapanewon
kapanewon_coords = {}
for kapanewon in unique_kapanewon:
    lat, lon = geocode_location(kapanewon)
    if lat and lon:
        kapanewon_coords[kapanewon] = (lat, lon)
```

```python
import folium
from folium.plugins import MarkerCluster

# Initialize the map centered around Yogyakarta
m = folium.Map(location=[-7.7956, 110.3695], zoom_start=10)

# Create a marker cluster
marker_cluster = MarkerCluster().add_to(m)

# Add each Kapanewon to the map
for kapanewon, coords in kapanewon_coords.items():
    folium.Marker(
        location=coords,
        popup=kapanewon,
        icon=folium.Icon(icon="cloud"),
    ).add_to(marker_cluster)

# Display the map
m
```

```python
df['Kabupaten'] = df['Location'].str.split(',').str[-2].str.strip()

# Manually specify the coordinates for the known Kabupaten values
manual_kabupaten_coords = {
    "sleman": (-7.7251, 110.4024),
    "bantul": (-7.9135, 110.4035),
    "yogyakarta": (-7.7956, 110.3695),
    "kulon progo": (-7.8262, 110.1662),
    "gunung kidul": (-8.0300, 110.6167)
}

# Count the number of properties for each Kabupaten
properties_by_kabupaten = df.groupby('Kabupaten').size().reset_index(name='Number_of_Properties')

# Calculate max_count and set max_radius
max_count = properties_by_kabupaten['Number_of_Properties'].max()
max_radius = 50

# Initialize the map centered around Yogyakarta
m = folium.Map(location=[-7.7956, 110.3695], zoom_start=10)

# Extract coordinates and properties count for each Kabupaten and add them to the map
for kabupaten, coords in manual_kabupaten_coords.items():
    count = properties_by_kabupaten[properties_by_kabupaten['Kabupaten'] == kabupaten]['Number_of_Properties'].values[0]
    
    # Calculate dynamic radius
    radius = (count / max_count) * max_radius
    
    popup_content = f"{kabupaten}<br>Properties: {count}"
    
    # Use circle markers with dynamic size
    folium.CircleMarker(
        location=coords,
        radius=radius,
        popup=popup_content,
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

# Display the map
m
```
